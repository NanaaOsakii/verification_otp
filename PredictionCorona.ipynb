```python
import PyPDF2
import pandas as pd

dates = ['19/05/2023', '12/05/2023', '05/05/2023', '28/04/2023', '21/04/2023',
         '14/04/2023', '07/04/2023', '31/03/2023', '24/03/2023', '17/03/2023',
         '10/03/2023', '06/03/2023', '05/03/2023', '04/03/2023', '03/03/2023',
         '02/03/2023', '01/03/2023', '28/02/2023', '27/02/2023', '26/02/2023',
         '25/02/2023', '24/02/2023', '23/02/2023', '22/02/2023', '21/02/2023',
         '20/02/2023', '19/02/2023', '18/02/2023', '17/02/2023', '16/02/2023',
         '15/02/2023', '14/02/2023', '13/02/2023', '12/02/2023', '11/02/2023',
         '10/02/2023', '09/02/2023', '08/02/2023', '07/02/2023', '06/02/2023',
         '05/02/2023', '04/02/2023', '03/02/2023', '02/02/2023', '01/02/2023',
         '31/01/2023', '30/01/2023', '29/01/2023', '28/01/2023', '27/01/2023',
         '26/01/2023', '25/01/2023', '24/01/2023', '23/01/2023', '22/01/2023',
         '21/01/2023', '20/01/2023', '19/01/2023', '18/01/2023', '17/01/2023',
         '16/01/2023', '15/01/2023', '14/01/2023', '13/01/2023', '12/01/2023',
         '11/01/2023', '10/01/2023', '09/01/2023', '08/01/2023', '07/01/2023',
         '06/01/2023', '05/01/2023', '04/01/2023', '03/01/2023', '02/01/2023',
         '01/01/2023']

def extract_cases_and_deaths(text):
    pattern = 'Béni Mellal - Khénifra'
    lines = text.split('\n')
    data = []
    has_data = False

    for line in lines:
        if pattern in line:
            values = line.split()
            if len(values) >= 6:
                cases = values[4]
                deaths = values[5]
                if cases.isdigit() and deaths.isdigit():
                    data.append({'Cases': cases, 'Deaths': deaths})
                    has_data = True

    if not has_data:
        data.append({'Cases': 0, 'Deaths': 0})

    return data

def extract_pdf_text(file_path):
    text = ''

    with open(file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        for page_number in range(num_pages):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

    return text

file_paths = ['/COVID-19.Morocco.13.19.05.2023.pdf','/COVID-19.Morocco.06-12.05.2023.pdf','/COVID-19.Morocco.29.04-05.05.2023.pdf','/COVID-19.Morocco.22-28.04.2023.pdf','/COVID-19.Morocco.15-21.04.2023.pdf','/COVID-19.Morocco.08-14.04.2023.pdf','/COVID-19.Morocco.01.07.04.2023.pdf','/COVID-19.Morocco.25-31.03.2023.pdf','/COVID-19.Morocco.18.24.03.2023.pdf','/COVID-19.Morocco.11-17.03.2023.pdf','/COVID-19.Morocco.06-10.03.2023.pdf',
'/06.03.2023.COVID-19.Morocco.pdf','/05.03.2023.COVID-19.Morocco.pdf','/04.03.2023.COVID-19.Morocco.pdf','/03.03.2023.COVID-19.Morocco.pdf','/02.03.2023.COVID-19.Morocco.pdf','/01.03.2023.COVID-19.Morocco.pdf','/28.02.2023.COVID-19.Morocco.pdf','/27.02.2023.COVID-19.Morocco.pdf','/26.02.2023.COVID-19.Morocco.pdf','/25.02.2023.COVID-19.Morocco.pdf','/24.02.2023.COVID-19.Morocco.pdf','/23.02.2023.COVID-19.Morocco.pdf','/22.02.2023.COVID-19.Morocco.pdf','/21.02.2023.COVID-19.Morocco.pdf',
'/20.02.2023.COVID-19.Morocco.pdf','/19.02.2023.COVID-19.Morocco.pdf','/18.02.2023.COVID-19.Morocco.pdf','/17.02.2023.COVID-19.Morocco.pdf','/16.02.2023.COVID-19.Morocco.pdf','/15.02.2023.COVID-19.Morocco.pdf','/14.02.2023.COVID-19.Morocco.pdf','/13.02.2023.COVID-19.Morocco.pdf','/12.02.2023.COVID-19.Morocco.pdf','/11.02.2023.COVID-19.Morocco.pdf','/10.02.2023.COVID-19.Morocco.pdf','/09.02.2023.COVID-19.Morocco.pdf','/08.02.2023.COVID-19.Morocco.pdf','/07.02.2023.COVID-19.Morocco.pdf','/06.02.2023.COVID-19.Morocco.pdf','/05.02.2023.COVID-19.Morocco.pdf','/04.02.2023.COVID-19.Morocco.pdf','/03.02.2023.COVID-19.Morocco.pdf','/02.02.2023.COVID-19.Morocco.pdf','/01.02.2023.COVID-19.Morocco.pdf','/31.01.2023.COVID-19.Morocco.pdf','/01.01.2023.COVID-19.Morocco.pdf','/29.01.2023.COVID-19.Morocco.pdf','/28.01.2023.COVID-19.Morocco.pdf','/27.01.2023.COVID-19.Morocco.pdf','/26.01.2023.COVID-19.Morocco.pdf','/25.01.2023.COVID-19.Morocco.pdf','/24.01.2023.COVID-19.Morocco.pdf','/23.01.2023.COVID-19.Morocco.pdf','/22.01.2023.COVID-19.Morocco.pdf','/21.01.2023.COVID-19.Morocco.pdf','/20.01.2023.COVID-19.Morocco.pdf','/19.01.2023.COVID-19.Morocco.pdf','/18.01.2023.COVID-19.Morocco.pdf','/17.01.2023.COVID-19.Morocco.pdf','/16.01.2023.COVID-19.Morocco.pdf','/15.01.2023.COVID-19.Morocco.pdf',
'/14.01.2023.COVID-19.Morocco.pdf','/13.01.2023.COVID-19.Morocco.pdf','/12.01.2023.COVID-19.Morocco.pdf','/11.01.2023.COVID-19.Morocco.pdf','/10.01.2023.COVID-19.Morocco.pdf','/09.01.2023.COVID-19.Morocco.pdf','/08.01.2023.COVID-19.Morocco.pdf','/07.01.2023.COVID-19.Morocco.pdf','/06.01.2023.COVID-19.Morocco.pdf','/05.01.2023.COVID-19.Morocco.pdf','/04.01.2023.COVID-19.Morocco.pdf','/03.01.2023.COVID-19.Morocco.pdf','/02.01.2023.COVID-19.Morocco.pdf','/01.01.2023.COVID-19.Morocco.pdf'
]

data = []

for i, file_path in enumerate(file_paths):
    pdf_text = extract_pdf_text(file_path)
    extracted_data = extract_cases_and_deaths(pdf_text)
    extracted_data[0]['Date'] = dates[i]  # Add the date to the extracted data
    data.extend(extracted_data)

df = pd.DataFrame(data)
print(df)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
df['Date'] = (df['Date'] - df['Date'].min()).dt.days

X = df[['Date']].values
y_cases = df['Cases'].values
y_deaths = df['Deaths'].values

X_train, X_test, y_cases_train, y_cases_test = train_test_split(X, y_cases, test_size=0.2, random_state=42)
X_train, X_test, y_deaths_train, y_deaths_test = train_test_split(X, y_deaths, test_size=0.2, random_state=42)

model_cases = LinearRegression()
model_cases.fit(X_train, y_cases_train)

model_deaths = LinearRegression()
model_deaths.fit(X_train, y_deaths_train)

next_months = np.array([[df['Date'].max() + i] for i in range(1, 7)])  # Prédire les 6 prochains mois
predicted_cases = model_cases.predict(next_months)
predicted_deaths = model_deaths.predict(next_months)

# Afficher les prédictions
for i in range(len(next_months)):
    print(f"Prévision pour le mois {i+1}: Cas - {predicted_cases[i]}, Décès - {predicted_deaths[i]}")
```
